```python
# pip install this if you did not yet.
from pandasgwas import get_studies
from pandasgwas.summary_statistics import search, browser, download, parse
```

```python
studies = get_studies(efo_trait = 'autism spectrum disorder')
studies.studies.head()
```

```python
len(studies.studies)
```

```python
for idx in range(len(studies.studies)):
    search_DF = search(PubMed_id=studies.studies["publicationInfo.pubmedId"].loc[idx],
                       study_accession_id=studies.studies["accessionId"].loc[idx])
    # some studies do not have data on the FTP server, so list only those which have
    if len(search_DF) > 0:
        display(search_DF.head())
```

```python
# select one study
search_DF = search(PubMed_id='33686288', study_accession_id='GCST90016596')
```

```python
# will display data tree in the browser
browser(search_DF)
```

```python
search_DF
```

```python
# this will downloaad around 200 MB 
download(search_DF)
```

```python
# and here is the data, this cell takes a bit of time to run (DF has 6 million lines)
df = parse(search_DF)
```
